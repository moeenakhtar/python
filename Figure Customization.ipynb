```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(1,9)
```

```python
y = np.arange(3,27,3)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1)
plt.tight_layout()
```

```python
print(type(axes))
```

```python
for ax in axes:
    print(ax.plot(x,y))
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(x,y)
axes[0].set_title("x-y graph")
axes[1].plot(x,x **2)
axes[1].set_title("x-x square graph")
plt.tight_layout()
```

```python
figure_s = plt.figure(figsize=(5,6), facecolor="red")
axes_s = figure_s.add_axes([0,0,1,1])
```

```python
figure_3, axes_30 = plt.subplots(nrows=1,ncols=3, figsize=(18,6))
figure_3.tight_layout()
```

```python
axes_30[0].plot(x,y,color="red")
axes_30[0].set_title("X-Y Graph")
axes_30[0].set_xlabel("X Values")
axes_30[0].set_ylabel("Y Values")

axes_30[1].plot(x,x**2,color="blue")
axes_30[1].set_title("X-x2 Graph")
axes_30[1].set_xlabel("X Values")
axes_30[1].set_ylabel("X^2 Values")

axes_30[2].plot(y,y**2,color="red")
axes_30[2].set_title("Y-Y^2 Graph")
axes_30[2].set_xlabel("Y Values")
axes_30[2].set_ylabel("Y^2 Values")
```

```python
figure_3
```

```python
figure_4, axes_40 = plt.subplots(figsize=(15,6))
axes_40 = figure_4.add_axes([0,0,1,1])
axes_40.plot(y,y**0.5,label="y square root", color="orange")
axes_40.plot(x,x**2,label="x square root", color="red")
axes_40.plot(x,y,label="x-y", color="blue")
axes_40.plot(x,np.sin(x),label="x-Sign", color="black")
axes_40.legend()
```
