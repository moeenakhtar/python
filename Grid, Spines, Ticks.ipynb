```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = np.arange(0,math.pi*2, 0.06)
```

```python
figure = plt.figure()
axes   = figure.add_axes([0,0,1,1])
y = np.sin(x)
axes.plot(x,y)
axes.set_xlabel("angle")
axes.set_title("Sine")
axes.set_xticks([0,2,4,6,8])
axes.set_xticklabels(["zero","Two","four","Six", "Eight"])
axes.set_yticks([-1,0,1])
axes.spines["right"].set_visible(False)
axes.spines["top"].set_visible(False)
axes.grid(color="b",ls="-", lw="0.35")
plt.show()
```
